```python
import pandas as pd
deliveries = pd.read_csv("deliveries.csv")
matches = pd.read_csv("matches.csv")
```

```python
unique_list = sharma_matches_ball_by_ball["match_id"].unique().tolist()
```

```python
len(unique_list)
```

```python
unique_matches = []
for id_match in unique_list:
    sharma_matches = matches[matches["id"]==id_match]
    unique_matches.append(sharma_matches.iloc[0])
```

```python
unique_matches
```

```python
unique_df = pd.DataFrame(unique_matches)
```

```python
unique_df
```

```python
unique_df["date"]=pd.to_datetime(unique_df["date"])
```

```python
unique_df
```

```python
unique_df = unique_df.sort_values(by="date")
```

```python
unique_df
```

```python
sharma_temp = deliveries[(deliveries["match_id"]==68)&(deliveries["batsman"]=="RG Sharma")]
```

```python
sharma_temp
```

```python
sharma_temp["batsman_runs"].sum()
```

```python
runs = []
for index,row in unique_df.iterrows():
    match_id = row["id"]
    sharma_temp = deliveries[(deliveries["match_id"]==match_id)&(deliveries["batsman"]=="RG Sharma")]
    sharma_runs = sharma_temp["batsman_runs"].sum()
    runs.append(sharma_runs)
```

```python
runs
```

```python
sum(runs)
```

```python
unique_df["runs"]=runs
```

```python
unique_df
```

```python
balls=[]
ten=[]
ten_thirty=[]
thirty_plus=[]
for index,row in unique_df.iterrows():
    match_id = row["id"]
    sharma_temp = deliveries[(deliveries["match_id"]==match_id)&(deliveries["batsman"]=="RG Sharma")]
    sharma_temp2 = sharma_temp[sharma_temp["wide_runs"]==0]
    sharma_balls = len(sharma_temp[sharma_temp["wide_runs"]==0])
    balls.append(sharma_balls)
    if sharma_balls<=10:
        ten.append(sharma_temp["batsman_runs"].sum())
        ten_thirty.append(0)
        thirty_plus.append(0)
    elif (sharma_balls>10) & (sharma_balls<=30):
        ten.append(sharma_temp2.iloc[0:10]["batsman_runs"].sum())
        ten_thirty.append(sharma_temp2.iloc[10:]["batsman_runs"].sum())
        thirty_plus.append(0)
    else:
        ten.append(sharma_temp2.iloc[0:10]["batsman_runs"].sum())
        ten_thirty.append(sharma_temp2.iloc[10:30]["batsman_runs"].sum())
        thirty_plus.append(sharma_temp2.iloc[30:]["batsman_runs"].sum())        
```

```python
balls
```

```python
ten
```

```python
ten_thirty
```

```python
thirty_plus
```

```python
unique_df["balls"]=balls
unique_df["ten"]=ten
unique_df["ten_thirty"]=ten_thirty
unique_df["thirty_plus"]=thirty_plus
```

```python
unique_df
```

```python
sr=[]
for index,row in unique_df.iterrows():
    sr_temp = row["runs"]/row["balls"]*100
    sr.append(sr_temp)
```

```python
sr
```

```python
unique_df["sr"]=sr
```

```python
unique_df
```
