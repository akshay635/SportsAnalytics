```python
unique_list = kohli_matches_ball_by_ball["match_id"].unique().tolist()
```

```python
unique_matches = []
for id_match in unique_list:
    kohli_matches = matches[matches["id"]==id_match]
    unique_matches.append(kohli_matches.iloc[0])
```

```python
unique_matches
```

```python
unique_df = pd.DataFrame(unique_matches)
```

```python
unique_df
```

```python
unique_df["date"] = pd.to_datetime(unique_df["date"])
```

```python
unique_df
```

```python
unique_df = unique_df.sort_values(by="date")
```

```python
unique_df
```

```python
kohli_temp = deliveries[(deliveries["match_id"]==63) & (deliveries["batsman"]=="V Kohli")]
```

```python
kohli_temp
```

```python
kohli_temp["batsman_runs"].sum()
```

```python
runs=[]
for index,row in unique_df.iterrows():
    match_id = row["id"]
    kohli_temp = deliveries[(deliveries["match_id"]==match_id) & (deliveries["batsman"]=="V Kohli")]
    kohli_runs = kohli_temp["batsman_runs"].sum()
    runs.append(kohli_runs)
```

```python
runs
```

```python
sum(runs)
```

```python
balls=[]
ten=[]
ten_thirty=[]
thirty_plus=[]
for index,row in unique_df.iterrows():
    match_id = row["id"]
    kohli_temp = deliveries[(deliveries["match_id"]==match_id) & (deliveries["batsman"]=="V Kohli")]
    kohli_temp2 = kohli_temp[kohli_temp["wide_runs"]==0]
    kohli_balls = len(kohli_temp[kohli_temp["wide_runs"]==0])
    balls.append(kohli_balls)
    if (kohli_balls<=10):
        ten.append(kohli_temp["batsman_runs"].sum())
        ten_thirty.append(0)
        thirty_plus.append(0)
    elif (kohli_balls>10) & (kohli_balls<=30):
        ten.append(kohli_temp2.iloc[0:10]["batsman_runs"].sum())
        ten_thirty.append(kohli_temp2.iloc[10:]["batsman_runs"].sum())
        thirty_plus.append(0)
    else:
        ten.append(kohli_temp2.iloc[0:10]["batsman_runs"].sum())
        ten_thirty.append(kohli_temp2.iloc[10:30]["batsman_runs"].sum())
        thirty_plus.append(kohli_temp2.iloc[30:]["batsman_runs"].sum())
```

```python
balls
```

```python
sum(balls)
```

```python
avg_sr = sum(runs)/sum(balls)*100
```

```python
avg_sr
```

```python
ten
```

```python
ten_thirty
```

```python
thirty_plus
```

```python
unique_df["runs"]=runs
unique_df["balls"]=balls
unique_df["ten"]=ten
unique_df["ten_thirty"]=ten_thirty
unique_df["thirty_plus"]=thirty_plus
```

```python
unique_df
```

```python
sr=[]
for index,row in unique_df.iterrows():
    sr_temp = row["runs"]/row["balls"]*100
    sr.append(sr_temp)
```

```python
sr
```

```python
unique_df["sr"]=sr
```

```python
unique_df
```

```python
kohli_group_season = unique_df.groupby(["season"])
```

```python
kohli_group_season
```

```python
kohli_data = kohli_group_season.mean()
```
